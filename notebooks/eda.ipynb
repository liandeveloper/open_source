## Bibliotecas

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns

sns.set_theme(style="whitegrid")
%matplotlib inline
```

```python
df = pd.read_csv('../datas/processed/repositories2.csv')
df["language"].value_counts().head(10).plot(kind="bar", figsize=(12,6))
plt.title("Ranking: Los Lenguajes de Programación Más Utilizados")
plt.xticks(rotation=45)
plt.show()
```
